# Notebook Example of LMQG
This is a colab notebook of [`lmqg`](https://github.com/asahi417/lm-question-generation#lmqg-language-model-for-question-generation-).

```python
from pprint import pprint
from lmqg import TransformersQG
```

## Question and Answer Generation (End2end)

```python
# initialize T5 model
model = TransformersQG(language='en', model='lmqg/t5-base-squad-qag')
```

```python
# paragraph to generate pairs of question and answer
context = "Financial risk encompasses potential loss or uncertainty in investments. It includes market, credit, operational, and liquidity risks. To mitigate these risks, diversify investments, conduct thorough assessments, use hedging techniques, maintain emergency funds, and comply with regulations. Effective risk management involves a combination of these methods."
question_answer = model.generate_qa(context)
pprint(question_answer)
```

## Question and Answer Generation (Multitask)


```python
# initialize model
model = TransformersQG(language='en', model='lmqg/t5-base-squad-qg-ae')
```

```python
# paragraph to generate pairs of question and answer
context = "Stakeholder analysis is a crucial component of effective project management and strategic planning. It involves identifying and assessing the interests, influence, and potential impact of individuals, groups, or organizations that are involved in or affected by a project or decision. This analysis helps in understanding the diverse perspectives, needs, and expectations of stakeholders, allowing for the development of strategies that can effectively engage, manage, and communicate with them. By mapping out the stakeholders, their roles, and their level of support or opposition, organizations can make informed decisions, mitigate risks, and enhance the likelihood of project success. It's a dynamic process that requires ongoing monitoring and adaptation as stakeholder dynamics can change over time. Ultimately, effective stakeholder analysis is a cornerstone of responsible and sustainable decision-making, as it ensures that the interests of all relevant parties are considered and addressed"
# model prediction
question_answer = model.generate_qa(context)
# the output is a list of tuple (question, answer)
```

## Question and Answer Generation (Pipeline)

```python
# initialize model
model = TransformersQG(model='lmqg/t5-base-squad-qg', model_ae='lmqg/t5-base-squad-ae')
```

```python
# paragraph to generate pairs of question and answer
context = "William Turner was an English painter who specialised in watercolour landscapes. He is often known as William Turner of Oxford or just Turner of Oxford to distinguish him from his contemporary, J. M. W. Turner. Many of Turner's paintings depicted the countryside around Oxford. One of his best known pictures is a view of the city of Oxford from Hinksey Hill."
# model prediction
question_answer = model.generate_qa(context)
# the output is a list of tuple (question, answer)
pprint(question_answer)
```

## Question Generation

```python
model = TransformersQG(language='en', model='lmqg/t5-base-squad-qg')
```

```python
# a list of paragraph
context = [
    "William Turner was an English painter who specialised in watercolour landscapes",
    "William Turner was an English painter who specialised in watercolour landscapes"
]
# a list of answer (same size as the context)
answer = [
    "William Turner",
    "English"
]

question = model.generate_q(list_context=context, list_answer=answer)
pprint(question)
```

## Answer Extraction

```python
# initialize model
model = TransformersQG(language='en', model='lmqg/t5-base-squad-ae')
```

```python
# model prediction
answer = model.generate_a("William Turner was an English painter who specialised in watercolour landscapes")
pprint(answer)
```
